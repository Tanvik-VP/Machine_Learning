```python
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
from sklearn.pipeline import Pipeline
import numpy as np

scaler = MinMaxScaler(feature_range=(0,22649))
encoder = LabelEncoder()

class data_transform(BaseEstimator,TransformerMixin):
    def __init__(self):
        pass
    def fit(x,y):
        return self
    def transform(self,x):
        x.Acreage = scaler.fit_transform(np.array(x.Acreage).reshape(-1,1))
        x.LandValue = scaler.fit_transform(np.array(x.LandValue).reshape(-1,1))
        x.BuildingValue = scaler.fit_transform(np.array(x.BuildingValue).reshape(-1,1))
        x.FinishedArea = scaler.fit_transform(np.array(x.FinishedArea).reshape(-1,1))   
        
        return x

data_pipeline = Pipeline([
    ('data pipeline',data_transform())
])
data_pipeline
```
